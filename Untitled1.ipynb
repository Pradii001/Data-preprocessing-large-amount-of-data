```python
df = df.dropna(subset=['Gender', 'Country', 'Degree', 'city', 'zip'], how='all')
```

```python
def clean_sponsors(value):
    try:
        return json.loads(value.replace("'", '"'))  # Convert string to list
    except:
        return []  # Return empty list if not valid JSON

df['PreferredSponsors'] = df['PreferredSponsors'].apply(clean_sponsors)
```

```python
df['zip'] = df['zip'].astype(str).str.replace(r'\D', '', regex=True)  # Remove non-numeric chars
df['zip'] = df['zip'].replace('', None)  # Convert empty strings to NaN
```

```python
df['isFromSocialMedia'] = df['isFromSocialMedia'].astype(str).str.lower().map({'true': True, 'false': False})
```

```python
df['Gender'].fillna('Unknown', inplace=True)
df['Country'].fillna('Unknown', inplace=True)
df['Degree'].fillna('Unknown', inplace=True)
df['city'].fillna('Unknown', inplace=True)

# Save cleaned data
cleaned_file_path = "/content/Cleaned_UserData.csv"
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned file saved: {cleaned_file_path}")

# Display a sample of cleaned data
df.head()
```

```python
cleaned_file_path = "/content/Cleaned_UserData.csv"
df = pd.read_csv(cleaned_file_path)
```

```python
# Fill missing values in 'zip' column with median (if any)
df['zip'] = df['zip'].fillna(df['zip'].median())

# Verify
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
categorical_columns = ['Gender', 'Country', 'Degree', 'city']

for col in categorical_columns:
    df[col] = encoder.fit_transform(df[col])

df.head()
```

```python
df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['zip']] = scaler.fit_transform(df[['zip']])

df.head()
```

```python
df['Sign Up Date'] = pd.to_datetime(df['Sign Up Date'])
df['signup_year'] = df['Sign Up Date'].dt.year
df['signup_month'] = df['Sign Up Date'].dt.month
df['signup_day'] = df['Sign Up Date'].dt.day

# Drop the original date column
df.drop('Sign Up Date', axis=1, inplace=True)

df.head()
```

```python
from scipy import stats
import numpy as np

# Remove rows where 'zip' is an extreme outlier
df = df[(np.abs(stats.zscore(df['zip'])) < 3)]

df.shape  # Check the number of remaining rows
```

```python
df.to_csv("/content/Preprocessed_UserData.csv", index=False)
print("Preprocessed file saved successfully!")
```
